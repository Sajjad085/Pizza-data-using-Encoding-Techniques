# ""Ordinal Encoder""

```python
df = pd.read_csv("Pizza Data.csv") #For to read the CSV file  
df.head()
```

```python
df.Size.unique() #For value counts using "unique" function
```

```python
from sklearn.preprocessing import OrdinalEncoder #import Ordianl Encoder
```

```python
pizza_sizes = ['Small (10")', 'Medium (12")', 'Large (14")', 'X-Large (16")',
       'Medium', 'Large', 'Personal', 'Small', 'Mini', 'Jumbo',
       'X Large (16")']
ordinal = OrdinalEncoder(categories=[pizza_sizes])
encoded = ordinal.fit_transform(df[['Size']])
```

```python
encoded
```

```python
encoded_size = pd.DataFrame(encoded,columns = ['Encoded'])
```

```python
df.Size = encoded_size
```

```python
df.Size
```

```python
cols = ['Size']
for col in cols: 
    uniques = df[col].unique()
    df[col] = OrdinalEncoder(categories=[uniques]).fit_transform(df[[col]])
```
